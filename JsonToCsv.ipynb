```python
#import library untuk mengolah data dalam format JSON
import json
import csv
```

```python
#menampilkan anatomi tweet
with open('2019-10-14_Tweets.json', 'r') as f:
    #mengambil tweet pada baris pertama
    text = f.readline()
    tweet = json.loads(text)
    #menampilkan twit dalam format json pada console
    print(json.dumps(tweet, indent=4))
```

```python
csvFile = open('2019-10-14_Tweets.csv', 'a')
csvWriter = csv.writer(csvFile)
```

```python
#menulis ke file csv
for l in open('2019-10-14_Tweets.json'):
    tw = json.loads(l)
    csvWriter.writerow([tw['id'],tw['created_at'],tw['full_text'].encode('utf-8')])
```

```python
csvFile.close()
```
